# Описание-задачи

Цель проекта — пройти полный цикл решения задачи машинного обучения: от изучения данных и определения постановки задачи до обучения моделей и оценки их качества. На первом этапе необходимо разобраться в предметной области, определить целевой признак и структуру данных, описать входные признаки и выявить зависимости между ними. Затем данные приводятся в корректный вид: уточняются типы значений, удаляются нерелевантные признаки, обрабатываются пропуски и кодируются категориальные значения.

После начальной подготовки проводится анализ распределений, выявляются возможные выбросы и аномалии, и данные очищаются от объектов, которые могут негативно повлиять на качество моделей. Далее исследуется взаимосвязь признаков, устраняется мультиколлинеарность, при необходимости выполняются преобразования и создаются новые признаки. Также анализируется масштаб данных и выполняется нормализация или скалирование, если это требуется.

На финальном этапе выбираются подходящие модели машинного обучения, определяются метрики качества, выполняется подбор оптимальных настроек, после чего модели обучаются и сравниваются между собой. По итоговым метрикам выбирается наилучший вариант решения задачи.

## 1.2. Первичный анализ данных

```python
def analyze_features(data):
    """Анализирует признаки датасета и генерирует таблицу с описанием"""
    
    feature_descriptions = []
    
    for column in data.columns:
        # Определяем тип данных
        dtype = str(data[column].dtype)
        unique_count = data[column].nunique()
        null_count = data[column].isnull().sum()
        
        # Анализируем специальные значения
        special_values = []
        if data[column].dtype == 'object':
            if (data[column] == '-').any():
                special_values.append('"Пропуск "-""')
            if (data[column] == 'Не определено').any():
                special_values.append('Пропуск - "Не определено"')
            if (data[column] == '').any() or (data[column] == ' ').any():
                special_values.append('Пропуск - пустые строки')
        
        # Определяем категорию признака
        if column == 'Гармония Бессмертия':
            feature_type = 'Целевой (непрерывный)'
        elif column == 'Номер пометки':
            feature_type = 'Идентификатор'
        elif 'Тип' in column or unique_count < 10:
            feature_type = 'Категориальный'
        elif dtype in ['object', 'string']:
            feature_type = 'Текстовый/Смешанный'
        elif dtype in ['float64', 'int64']:
            if unique_count > 50:
                feature_type = 'Непрерывный'
            else:
                feature_type = 'Дискретный'
        else:
            feature_type = 'Неизвестный'
        
        # Статистика для числовых признаков
        if pd.api.types.is_numeric_dtype(data[column]):
            min_val = data[column].min()
            max_val = data[column].max()
            range_info = f"{min_val:.3f}-{max_val:.3f}"
        else:
            range_info = f"{unique_count} уникальных значений"
        
        # Особенности
        features = []
        if null_count > 0:
            features.append(f"пропуски: {null_count}")
        if special_values:
            features.extend(special_values)
        if unique_count == 1:
            features.append("константа")
        if unique_count == len(data):
            features.append("уникальные значения")
        
        feature_info = {
            'Признак': column,
            'Тип данных': dtype,
            'Категория': feature_type,
            'Уникальных значений': unique_count,
            'Диапазон/Значения': range_info,
            'Особенности': ', '.join(features) if features else '-'
        }
        
        feature_descriptions.append(feature_info)
    
    return pd.DataFrame(feature_descriptions)

display(analyze_features(df))
```

## 1.4. Кодирование категориальных признаков

Применён OrdinalEncoder, т.к. единственный категориальный признак это Тип Вектора Мощи, который имеет отношение порядка.

```python
from sklearn.preprocessing import OrdinalEncoder

m = ['Тип Вектора Мощи']

# Задаем порядок
categories = [['Слабый', 'Ниже среднего', 'Выше среднего', 'Сильный']]

# Сохраняем оригинальные данные для проверки
X_train_original = X_train.copy()
X_test_original = X_test.copy()

encoder = OrdinalEncoder(categories=categories)

for i in m:
    X_train[i] = encoder.fit_transform(X_train[[i]]).ravel()
    X_test[i] = encoder.transform(X_test[[i]]).ravel()

def check_order_preservation(original_series, encoded_series, feature_name, expected_categories):
    print(f" Проверка для признака: {feature_name}")
    
    # Получаем уникальные значения и их коды
    unique_original = original_series.unique()
    mapping = {}
    
    for val in unique_original:
        # Берем первое значение для каждой категории
        encoded_val = encoded_series[original_series == val].iloc[0]
        mapping[val] = encoded_val
    
    print("Маппинг категорий:")
    for orig, encoded in mapping.items():
        print(f"  {orig} -> {encoded}")
    
    print("\nПроверка соответствия ожидаемому порядку:")
    for i, category in enumerate(expected_categories[0]):  # т.к. categories = [['...']]
        print(f"  {category}: ожидаемый код - {i}, фактический код - {mapping[category]}")
    
    return mapping

# Применяем проверку для каждого признака
for feature in m:
    mapping = check_order_preservation(X_train_original[feature], X_train[feature], feature, categories)
```

## 1.5. Преобразование типов данных

Все что не получается преобразовать в число, т.е. в float или int преобразуется в nan (нужно т.к. в датасете есть пропуски: - и Не определено)

```python
def transformation(t):
    numeric_columns = ['Вектор Мощи', 'Скорость перехода через портал', 'Приток Силы Потока', 
                  'Ритм магического ядра', 'Поток Энергий', 'Сила Левого Потока', 
                  'Сила Правого Потока', 'Пламя Стихий',
                  'Температура выдоха Истока', 'Приток давления Выдоха Истока', 'Давление выдоха Истока', 'Древний Ветер',
                  'Печать Чародея', 'Эмульсия Истока', 'Дыхание Истока']

    # Все что не получается преобразовать в число, т.е. в float или int преобразуется в nan (нужно т.к. в датасете есть пропуски: - и Не определено)
    for col in numeric_columns:
        t[col] = pd.to_numeric(t[col], errors='coerce')
    object_columns = df.select_dtypes(include=['object']).columns
    print(f"Столбцы для преобразования из object в float: {list(object_columns)}")
    
    for col in object_columns:
        # Заменяем специальные значения на NaN перед преобразованием
        t[col] = t[col].replace(['Не определено', '-', ' ', ''], np.nan)
        # Преобразуем в числовой формат
        t[col] = pd.to_numeric(t[col], errors='coerce')
        
    print("Преобразование завершено")
    print(f"Типы данных после преобразования:")
    print(t.dtypes)
    
transformation(X_train)

transformation(X_test)
```

## 1.6. Обработка вещественных признаков (заполнение пропусков)

Статистический подход

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp, entropy
from scipy import stats

# Функция для вычисления KL-дивергенции
def calculate_kl_divergence(original_data, imputed_data, feature, bins=30):
    # Убираем NaN значения из оригинальных данных
    original_clean = original_data[feature].dropna()
    imputed_clean = imputed_data[feature]
    
    # Проверяем, что есть данные для сравнения
    if len(original_clean) == 0 or len(imputed_clean) == 0:
        return np.nan
    
    # Создаем гистограммы с одинаковыми бинами
    min_val = min(original_clean.min(), imputed_clean.min())
    max_val = max(original_clean.max(), imputed_clean.max())
    
    # Создаем бины
    bin_edges = np.linspace(min_val, max_val, bins + 1)
    
    # Вычисляем гистограммы
    hist_original, _ = np.histogram(original_clean, bins=bin_edges, density=True)
    hist_imputed, _ = np.histogram(imputed_clean, bins=bin_edges, density=True)
    
    # Добавляем небольшое значение чтобы избежать нулей
    hist_original = hist_original + 1e-10
    hist_imputed = hist_imputed + 1e-10
    
    # Нормализуем
    hist_original = hist_original / np.sum(hist_original)
    hist_imputed = hist_imputed / np.sum(hist_imputed)
    
    # Вычисляем KL-дивергенцию
    kl_div = entropy(hist_original, hist_imputed)
    
    return kl_div  # Чем ниже KL-дивергенция, тем более схожи распределения

# Функция для визуализации распределений всех методов
def plot_all_methods_comparison(original_data, imputed_data_dict, feature):
    n_methods = len(imputed_data_dict)
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()
    
    # Оригинальное распределение (без NaN)
    original_clean = original_data[feature].dropna()
    
    # Первый график - оригинальное распределение
    axes[0].hist(original_clean, bins=30, alpha=0.7, color='blue', density=True)
    axes[0].set_title(f'{feature}\nОригинальное распределение\n(без пропусков)')
    axes[0].set_xlabel('Значение')
    axes[0].set_ylabel('Плотность')
    
    # Графики для каждого метода
    for idx, (method_name, imputed_data) in enumerate(imputed_data_dict.items(), 1):
        if idx < len(axes):
            axes[idx].hist(imputed_data[feature], bins=30, alpha=0.7, color='red', density=True)
            
            # Вычисляем метрики для отображения
            ks_stat, ks_pval = ks_2samp(original_clean, imputed_data[feature])
            kl_div = calculate_kl_divergence(original_data, imputed_data, feature)
            
            axes[idx].set_title(f'{method_name}\n'
                              f'KS: {ks_stat:.3f} (p: {ks_pval:.3f})\n'
                              f'KL: {kl_div:.3f}')
            axes[idx].set_xlabel('Значение')
            axes[idx].set_ylabel('Плотность')
    
    # Скрываем лишние subplot'ы если есть
    for idx in range(n_methods + 1, len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()

# Определяем признаки с пропусками
missing_features = X_train.columns[X_train.isnull().any()].tolist()
print(f"Признаки с пропусками: {missing_features}")
print(f"Количество пропусков по признакам:")
print(X_train[missing_features].isnull().sum())

# Способ 1: Статистические подходы
print("Статистические подходы")

# Для каждого признака тестируем разные стратегии заполнения
best_imputations_stats = {}

for feature in missing_features:
    print(f" Анализ признака: {feature}")
    
    # Возможные значения для заполнения
    test_values = [
        ('mean', X_train[feature].mean()),
        ('median', X_train[feature].median()),
        ('mode', X_train[feature].mode()[0]),
        ('min', X_train[feature].min()),
        ('max', X_train[feature].max())
    ]
    
    best_score = float('inf')  # Для KL-дивергенции чем меньше, тем лучше
    best_value = None
    best_strategy = None
    
    # Создаем словарь для хранения всех методов для визуализации
    all_methods_data = {}
    
    # Тестируем каждое значение
    for strategy, value in test_values:
        # Создаем копию данных для тестирования
        temp_data = X_train.copy()
        temp_data[feature] = temp_data[feature].fillna(value)
        
        # Сохраняем данные для визуализации
        all_methods_data[strategy] = temp_data
        
        # Сравниваем распределения с помощью KL-дивергенции
        score = calculate_kl_divergence(X_train, temp_data, feature)
        
        # Также вычисляем KS-тест для информации
        ks_statistic, ks_pvalue = ks_2samp(X_train[feature].dropna(), temp_data[feature])
        
        print(f"  {strategy}: {value:.3f}, KL-дивергенция: {score:.4f}, KS p-value: {ks_pvalue:.4f}")
        
        # Сохраняем лучшее значение (чем ниже KL-дивергенция, тем лучше)
        if score < best_score:
            best_score = score
            best_value = value
            best_strategy = strategy

    # Визуализируем все методы
    print(f"  Визуализация всех методов для {feature}:")
    plot_all_methods_comparison(X_train, all_methods_data, feature)
    
    # Сохраняем лучшую стратегию
    best_imputations_stats[feature] = {
        'strategy': best_strategy,
        'value': best_value,
        'kl_score': best_score
    }
    
    print(f"Лучшая стратегия для {feature}: {best_strategy} = {best_value:.3f} (KL-дивергенция: {best_score:.4f})")

# Применяем лучшие статистические стратегии к обучающим данным
X_train_statistical = X_train.copy()
for feature, params in best_imputations_stats.items():
    X_train_statistical[feature] = X_train_statistical[feature].fillna(params['value'])

```

Количество пропусков в признакх Температура выдоха Истока и Древний ветер чрезвычайно велико, более 80%, что говорит о том, что эти столбцы можно удалить, т.к. никакое заполнение не сможет дать нам хорошие данные.

```python
# Сравнение методов для каждого признака

comparison_results = []

for feature in missing_features:
    print(f"Сравнение методов для признака: {feature}")
    
    # Собираем все заполненные версии
    filled_versions = {
        'Statistical': X_train_statistical,
        'KNN': X_train_knn,
        'Iterative': X_train_iterative,
        'Simple_median': X_train_simple
    }
    
    # Вычисляем KS-статистику для каждого метода
    for method_name, filled_data in filled_versions.items():
        score = calculate_kl_divergence(X_train, filled_data, feature)
        
        if not np.isnan(score):
            comparison_results.append({
                'Feature': feature,
                'Method': method_name,
                'KS_Score': score
            })
            print(f"  {method_name}: KS p-value = {score:.4f}")
        else:
            print(f"  {method_name}: невозможно вычислить")

# Создаем таблицу сравнения
comparison_df = pd.DataFrame(comparison_results)
pivot_comparison = comparison_df.pivot(index='Feature', columns='Method', values='KS_Score')

print("\nKS p-value по методам и признакам (чем выше, тем лучше):")
display(pivot_comparison)

# Выбираем лучший метод для каждого признака
best_methods = {}
for feature in missing_features:
    feature_comparison = comparison_df[comparison_df['Feature'] == feature].dropna()
    if not feature_comparison.empty:
        best_method = feature_comparison.loc[feature_comparison['KS_Score'].idxmax()]  # максимизируем p-value
        best_methods[feature] = best_method['Method']
        
        print(f"Лучший метод для {feature}: {best_method['Method']} (KS p-value: {best_method['KS_Score']:.4f})")

# Выбираем общий лучший метод (по среднему KS score)
avg_ks = comparison_df.groupby('Method')['KS_Score'].mean()
best_overall_method = avg_ks.idxmax()
print(f"\nОбщий лучший метод: {best_overall_method} (средний KS p-value: {avg_ks[best_overall_method]:.4f})")

# Используем лучший метод как основной
if best_overall_method == 'Statistical':
    X_train_imputed = X_train_statistical.copy()
    best_imputer = None  # для статистического метода нет imputer'а
elif best_overall_method == 'KNN':
    X_train_imputed = X_train_knn.copy()
    best_imputer = knn_imputer
elif best_overall_method == 'Iterative':
    X_train_imputed = X_train_iterative.copy()
    best_imputer = iterative_imputer
else:  # Simple_median
    X_train_imputed = X_train_simple.copy()
    best_imputer = simple_imputer

print(f"Выбран основной метод заполнения: {best_overall_method}")
```

```python
def calculate_nrmse(original, imputed, mask):
    # Извлекаем только маскированные значения (искусственно созданные пропуски)
    original_values = original[mask]
    imputed_values = imputed[mask]
    
    # Проверяем, что есть данные для сравнения
    if len(original_values) == 0:
        return np.nan
        
    # Вычисляем RMSE
    rmse = np.sqrt(mean_squared_error(original_values, imputed_values))
    
    # Нормализуем по диапазону значений
    value_range = original_values.max() - original_values.min()
    
    # Избегаем деления на 0
    if value_range == 0:
        return rmse
    else:
        return rmse / value_range

def evaluate_imputation_quality(feature_data, missing_rate=0.2, n_trials=3):
    nrmse_scores = []
    
    for trial in range(n_trials):
        # Создаем копию чистых данных
        data_clean = feature_data.dropna().values
        
        if len(data_clean) == 0:
            continue
            
        # Создаем случайную маску пропусков
        np.random.seed(SEED + trial)
        n_samples = len(data_clean)
        mask = np.random.random(size=n_samples) < missing_rate
        
        # Пропускаем если нет маскированных значений
        if mask.sum() == 0:
            continue
            
        # Искусственно создаем пропуски
        data_with_nans = data_clean.copy()
        data_with_nans[mask] = np.nan
        
        # Заполняем медианой (простой метод для оценки)
        from sklearn.impute import SimpleImputer
        imputer = SimpleImputer(strategy='median')
        data_imputed = imputer.fit_transform(data_with_nans.reshape(-1, 1)).flatten()
        
        # Вычисляем NRMSE
        nrmse = calculate_nrmse(data_clean, data_imputed, mask)
        
        if not np.isnan(nrmse):
            nrmse_scores.append(nrmse)

    
    return np.mean(nrmse_scores) if nrmse_scores else np.nan

# Создаем "чистый" датасет (убираем строки с пропусками)
clean_mask = ~X_train.isnull().any(axis=1)
X_clean = X_train[clean_mask].copy()

print(f"Размер чистого датасета для оценки: {len(X_clean)} строк")

# Оцениваем качество для каждого признака с пропусками
print("Оценка качества заполнения (NRMSE):")
nrmse_results = []

for feature in missing_features:
    feature_data = X_clean[feature]
    nrmse_score = evaluate_imputation_quality(feature_data)
    
    quality = 'Отличное' if nrmse_score < 0.05 else 'Хорошее' if nrmse_score < 0.1 else 'Удовлетворительное' if nrmse_score < 0.2 else 'Плохое'
    nrmse_results.append({
        'Feature': feature,
        'NRMSE': nrmse_score,
        'Quality': quality
    })
    print(f"  {feature}: NRMSE = {nrmse_score:.4f} ({quality})")

nrmse_df = pd.DataFrame(nrmse_results)
print("Итоговая оценка качества заполнения:")
display(nrmse_df)

# Визуализация качества заполнения
plt.figure(figsize=(12, 6))
colors = ['green' if x < 0.05 else 'lightgreen' if x < 0.1 else 'orange' if x < 0.2 else 'red' for x in nrmse_df['NRMSE']]
bars = plt.bar(nrmse_df['Feature'], nrmse_df['NRMSE'], color=colors, alpha=0.7)

plt.axhline(y=0.05, color='darkgreen', linestyle='--', label='Отличное качество (NRMSE < 0.05)')
plt.axhline(y=0.1, color='green', linestyle='--', label='Хорошее качество (NRMSE < 0.1)')
plt.axhline(y=0.2, color='orange', linestyle='--', label='Удовлетворительное качество (NRMSE < 0.2)')

plt.title('Оценка качества заполнения пропусков (NRMSE)')
plt.xlabel('Признак')
plt.ylabel('NRMSE')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

## 1.7. Обогащение пространства признаков

По предложенным специалистами формулам:

* Нельзя испольозвать для решения задачи (поскольку содержат информацию о целевом признаке):

* Степень износа = Дыхание Истока / Гармония Бессмертия

* Расхождение стабильности = abs(Дыхание Истока - Гармония Бессмертия)

* Баланс угасания = (Дыхание Истока - Гармония Бессмертия) / Скорость перехода

Можно использовать (не содержат целевую переменную):

* Совокупная мощность = Сила Левого Потока + Сила Правого Потока

* Суммарная сила потоков = Сила Левого Потока + Сила Правого Потока + Приток Силы Потока

* Общая сила ядра = Ритм магического ядра × Приток Силы Потока

* Общее давление выхода = Приток давления Выдоха Истока + Давление выдоха Истока

* Магическая производительность = Скорость перехода / Эмульсия Истока

* Эффективность ядра = Общая сила ядра / Эмульсия Истока

* Разница температур = Пламя стихии - Температура Вдоха Истока

```python
def add_feat(t):
    # Добавление признаков по совету 
    # Суммарная мощность потоков
    t['Суммарная_сила_потоков'] = t['Сила Левого Потока'] + t['Сила Правого Потока']
    t['Общая_сила_всех_потоков'] = t['Сила Левого Потока'] + t['Сила Правого Потока'] + t['Приток Силы Потока']
    
    # Общая сила ядра
    t['Общая_сила_ядра'] = t['Ритм магического ядра'] * t['Приток Силы Потока']
    
    # Общее давление на выходе
    t['Общее_давление_выхода'] = t['Приток давления Выдоха Истока'] + t['Давление выдоха Истока']
    
    # Магическая производительность
    t['Магическая_производительность'] = t['Скорость перехода через портал'] / t['Эмульсия Истока']
    
    # Эффективность ядра
    t['Эффективность_ядра'] = t['Общая_сила_ядра'] / t['Эмульсия Истока']
    
    # Разница температур для магической мощности
    t['Разница_температур'] = t['Пламя Стихий'] - 736 #736 - единственное значение столбца Температура вдоха Истока
    
    print(f"Размер данных после создания новых признаков: {t.shape}")
    
add_feat(X_train)
add_feat(X_test)
```

## 1.8. Детекция выбросов и аномалий в данных

Поиск выбросов с помощью IQR и Robust Z-score

```python
def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers.index.tolist()

# Функция для поиска выбросов методом Robust Z-score (коэф - 2.5)
def find_outliers_robust_zscore(data, column, threshold=2.5):
    median = data[column].median()
    mad = np.median(np.abs(data[column] - median))
    if mad == 0:
        mad = 1e-6  # Избегаем деления на ноль
    robust_z = 0.6745 * (data[column] - median) / mad
    outliers = data[np.abs(robust_z) > threshold]
    return outliers.index.tolist()

# Построение ящиков с усами для каждого столбца
numeric_columns = X_train.select_dtypes(include=[np.number]).columns
numeric_columns = numeric_columns.drop('Гармония Бессмертия') if 'Гармония Бессмертия' in numeric_columns else numeric_columns

print(f"Анализируем {len(numeric_columns)} числовых признаков:")

# Словари для хранения выбросов
iqr_outliers_dict = {}
robust_outliers_dict = {}
common_outliers_dict = {}

# Анализ каждого признака
for i, column in enumerate(numeric_columns):
    print(f"Признак {i+1}/{len(numeric_columns)}: {column}")
    
    # Построение boxplot
    plt.figure(figsize=(10, 6))
    plt.subplot(1, 2, 1)
    sns.boxplot(y=X_train[column])
    plt.title(f'Boxplot: {column}')
    
    plt.subplot(1, 2, 2)
    plt.hist(X_train[column], bins=50, alpha=0.7, edgecolor='black')
    plt.title(f'Histogram: {column}')
    plt.tight_layout()
    plt.show()
    
    # Поиск выбросов
    iqr_outliers = find_outliers_iqr(X_train, column)
    robust_outliers = find_outliers_robust_zscore(X_train, column, threshold=2.5)
    common_outliers = list(set(iqr_outliers) & set(robust_outliers))
    
    iqr_outliers_dict[column] = iqr_outliers
    robust_outliers_dict[column] = robust_outliers
    common_outliers_dict[column] = common_outliers
    
    print(f"Выбросы IQR: {len(iqr_outliers)}")
    print(f"Выбросы Robust Z-score: {len(robust_outliers)}")
    print(f"Общие выбросы (пересечение): {len(common_outliers)}")

# Собираем все уникальные строки с выбросами
all_iqr_outliers = set()
all_robust_outliers = set()
all_common_outliers = set()

for column in numeric_columns:
    all_iqr_outliers.update(iqr_outliers_dict[column])
    all_robust_outliers.update(robust_outliers_dict[column])
    all_common_outliers.update(common_outliers_dict[column])

print("Поиск выбросов:")
print(f"Всего строк с выбросами IQR: {len(all_iqr_outliers)}")
print(f"Всего строк с выбросами Robust Z-score: {len(all_robust_outliers)}")
print(f"Всего общих выбросов (пересечение): {len(all_common_outliers)}")
```

## 2.2. Исправление проблемы мультиколлинеарности в данных

```python
def eliminate_multicollinearity(X_train, X_test):
    """
    Устранение мультиколлинеарности на основе знаний предметной области
    """
    # Создаем копии данных
    X_train_clean = X_train.copy()
    X_test_clean = X_test.copy()
    
    # Коэффициент эффективности давления
    X_train_clean['Коэф_эффективности_давления'] = (
        X_train_clean['Приток давления Выдоха Истока'] / 
        (X_train_clean['Давление выдоха Истока'])
    )
    X_test_clean['Коэф_эффективности_давления'] = (
        X_test_clean['Приток давления Выдоха Истока'] / 
        (X_test_clean['Давление выдоха Истока'])
    )
    
    # Отношение температуры к пламени стихий
    X_train_clean['Температурный_коэффициент'] = (
        X_train_clean['Температура выдоха Истока'] / 
        (X_train_clean['Пламя Стихий'])
    )
    X_test_clean['Температурный_коэффициент'] = (
        X_test_clean['Температура выдоха Истока'] / 
        (X_test_clean['Пламя Стихий'])
    )
    
    # Удаляем некоторые сильно коррелирующие признаки (в крайнем случае)
    columns_to_drop = [
        'Сила Правого Потока',
        'Сила Левого Потока',
        'Приток давления Выдоха Истока',
        'Давление выдоха Истока',
        'Пламя Стихий',
        'Приток Силы Потока',
        'Скорость перехода через портал',
        'Эмульсия Истока',
        'Ритм магического ядра',
        'Температура выдоха Истока'
    ]
    
    X_train_clean = X_train_clean.drop(columns=columns_to_drop, errors='ignore')
    X_test_clean = X_test_clean.drop(columns=columns_to_drop, errors='ignore')
    
    print(f"Размерность после устранения мультиколлинеарности: {X_train_clean.shape}")
    
    return X_train_clean, X_test_clean

# Применяем функцию устранения мультиколлинеарности
X_train_clean, X_test_clean = eliminate_multicollinearity(X_train, X_test)
```

**Описание работы моделей**

--- 
Ridge Regressor
Это регуляризованная версия линейной регрессии. Модель строит линейную зависимость между признаками и целевой переменной (y = w1*x1 + w2*x2 + ... + b), но добавляет в функцию потерь штраф (L2-регуляризацию) за слишком большие коэффициенты w.

Стандартная линейная регрессия минимизирует сумму квадратов ошибок (MSE).

Ridge добавляет к этой сумме штрафное слагаемое — сумму квадратов всех коэффициентов, умноженную на гиперпараметр alpha.

Цель: Минимизировать MSE + alpha * (w1² + w2² + ...).

Параметр alpha контролирует силу регуляризации:

alpha = 0 -> обычная линейная регрессия (может переобучаться).

alpha > 0 -> коэффициенты "сжимаются" к нулю, но никогда не становятся точно нулевыми. Это снижает дисперсию модели, повышает устойчивость к мультиколлинеарности (когда признаки коррелируют друг с другом) и борется с переобучением.

---
Random Forest Regressor	

Баггинг (Bootstrap Aggregating): Создается много (например, 100) случайных подвыборок (бутстрап) из исходных данных (с повторениями).

Случайность в признаках: Для каждого узла в каждом дереве при поиске лучшего разделения рассматривается не все признаки, а лишь их случайное подмножество (например, sqrt(n_features)).

Построение деревьев: На каждой подвыборке строится отдельное дерево решений без сильного ограничения глубины (часто до полного разделения выборки).

Агрегация предсказаний: Для получения итогового прогноза модели усредняются предсказания всех деревьев леса. Это снижает дисперсию, которую имеет одно дерево.

---
XGBoost Regressor
Последовательное построение: Модель строится итеративно (например, 100 итераций = 100 деревьев).

Исправление ошибок: Первое дерево делает предсказание. Второе дерево обучается не на исходные значения y, а на градиент (производную) функции потерь относительно предсказания первого дерева (фактически — на остатки с учетом сложности функции потерь). Третье — на остатки первых двух, и так далее.

Регуляризация: В отличие от простого бустинга, XGBoost включает L1 (Lasso) и L2 (Ridge) регуляризацию непосредственно в целевую функцию, которую оптимизирует, что позволяет контролировать сложность деревьев и бороться с переобучением.

Другие оптимизации: Использует вторые производные (гессианы) для более точного шага, эффективные структуры данных, распараллеливание, методы борьбы с пропущенными значениями.

---
CatBoost Regressor
Особенности для категориальных признаков: Вместо классического One-Hot Encoding или Label Encoding, CatBoost использует ordered encoding (упорядоченное кодирование) на основе целевой переменной, что решает проблему "утечки данных" (data leakage) при кодировании.

Симметричные деревья: Строит сбалансированные (симметричные) деревья, где на каждом уровне все узлы используют один и тот же признак для разделения. Это ускоряет предсказание и уменьшает переобучение.

Усовершенствованный градиентный бустинг: Использует модифицированный алгоритм для вычисления градиентов, который также борется со смещением (bias), возникающим при последовательном обучении.

---
KNeighbors Regressor (KNN)
Обучение: Модель просто сохраняет всю обучающую выборку (X_train, y_train).

Предсказание: Для нового объекта:

Вычисляется расстояние (обычно евклидово, манхэттенское и др.) от него до всех точек в X_train.

Находится k объектов с наименьшим расстоянием — "ближайшие соседи".

Итоговый прогноз — это среднее арифметическое значений y_train этих соседей (для регрессии). Для классификации — мода.

---
SVR (Support Vector Regressor)
Задается параметр e (epsilon-tube): Строится "трубка" шириной 2*e вокруг предполагаемой гиперплоскости.

Цель: Найти такую гиперплоскость, чтобы максимальное количество точек обучающих данных лежало внутри этой трубки. Ошибкой считаются только те точки, которые выходят за пределы трубки.

Ядра (kernels): Как и в SVM, с помощью ядерного трюка можно строить нелинейную регрессию, проецируя данные в пространство более высокой размерности, где линейная гиперплоскость может описать сложную зависимость (используются RBF, полиномиальные и другие ядра).

Параметр C: Контролирует компромисс между допустимой шириной трубки (e) и штрафом за точки, выходящие за ее пределы. Малое C — широкая трубка, возможны большие ошибки. Большое C — уже трубка, модель старается учесть каждую точку (риск переобучения).

**Используемые метрики**

---
RMSE (Root Mean Squared Error) - Среднеквадратичная ошибка

RMSE = sqrt((1/n)*E(y_i-y^_i)^2)

n - количество наблюдений

y_i - истинное значение целевой переменной

y^_i - предсказанное значение

Интерпретация:
Cредняя ожидаемая разница +/- между прогнозным значением и фактическим значением.
Показывает стандартное отклонение ошибок предсказания.
Измеряется в тех же единицах, что и целевая переменная.
Более чувствительна к большим ошибкам (выбросам) из-за возведения в квадрат.

---
MAE (Mean Absolute Error) - Средняя абсолютная ошибка

MAE = (1/n)*E|(y_i-y^_i)|

Интерпретация:

Средняя величина абсолютных ошибок предсказания.

Менее чувствительна к выбросам по сравнению с RMSE.

Легче интерпретировать: "в среднем модель ошибается на X единиц".

--- 
R^2 (Коэффициент детерминации)

R² = 1 - E(y_i-y^_i)^2/E(y_i-y^-)^2

y^- - среднее значение целевой переменной

Знаменатель представляет собой общую дисперсию целевой переменной

Интерпретация:

Показывает долю дисперсии целевой переменной, объясненную моделью

---
Итог:

RMSE — главная метрика, потому что нужна строгая оценка больших ошибок.

MAE — показывает базовую среднюю ошибку, полезно при шумных данных.

R² — показывает, нашла ли модель хоть какую-то структуру при корреляции 0.03.

# 4. Обучение моделей ML и подбор гиперпараметров

```python
DATA_VERSIONS = {
    'STD': (X_train_std, X_test_std),
    'MM': (X_train_mm, X_test_mm),
    'ROBUST': (X_train_robust, X_test_robust),
    'PCA': (X_train_pca, X_test_pca)
}

BEST_RESULTS = {}

TARGET_METRIC = 'RMSE'

MODELS = {
    'Ridge': Ridge,
    'RandomForest': RandomForestRegressor,
    'XGBoost': xgb.XGBRegressor,
    'CatBoost': cb.CatBoostRegressor,
    'KNN': KNeighborsRegressor,
    'SVR': SVR
}

def objective(trial, model_name, X_train_data, y_train_data):
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    rmse_scores = []

    # Получаем сетку гиперпараметров
    grid = HYPERPARAM_GRIDS[model_name]
    params = {}

    for param, cfg in grid.items():

        if cfg[0] == 'int':
            params[param] = trial.suggest_int(param, cfg[1], cfg[2])

        elif cfg[0] == 'float':
            if len(cfg) == 4 and cfg[3] == 'log':
                params[param] = trial.suggest_float(param, cfg[1], cfg[2], log=True)
            else:
                params[param] = trial.suggest_float(param, cfg[1], cfg[2])

        elif cfg[0] == 'categorical':
            params[param] = trial.suggest_categorical(param, cfg[1])

    # Инициализация модели
    if model_name in ['RandomForest', 'XGBoost']:
        model = MODELS[model_name](random_state=42, n_jobs=-1, **params)

    elif model_name == 'CatBoost':
        model = MODELS[model_name](verbose=0, random_state=42, **params)

    elif model_name == 'KNN':
        model = MODELS[model_name](n_jobs=-1, **params)

    else:
        model = MODELS[model_name](**params)

    # Кросс-валидация
    for train_idx, val_idx in kf.split(X_train_data):
        X_train_fold, X_val_fold = X_train_data.iloc[train_idx], X_train_data.iloc[val_idx]
        y_train_fold, y_val_fold = y_train_data.iloc[train_idx], y_train_data.iloc[val_idx]

        model.fit(X_train_fold, y_train_fold)
        preds = model.predict(X_val_fold)
        rmse_scores.append(np.sqrt(mean_squared_error(y_val_fold, preds)))

    return np.mean(rmse_scores)

N_TRIALS = 5

for model_name, ModelClass in MODELS.items():
    for data_version, (X_train_data, X_test_data) in DATA_VERSIONS.items():

        print(f"\nОптимизация: {model_name} на данных {data_version}...")

        # Создаём study
        study = optuna.create_study(
            direction='minimize',
            study_name=f'{model_name}_{data_version}'
        )

        # Objective-функция
        objective_func = lambda trial: objective(
            trial, model_name, X_train_data, y_train
        )

        # Запуск оптимизации
        study.optimize(objective_func, n_trials=N_TRIALS, show_progress_bar=False)

        best_params = study.best_params
        best_cv_rmse = study.best_value

        # Обучение финальной модели
        if model_name in ['RandomForest', 'XGBoost']:
            final_model = ModelClass(random_state=42, n_jobs=-1, **best_params)
        elif model_name == 'KNN':
            final_model = ModelClass(n_jobs=-1, **best_params)
        elif model_name == 'CatBoost':
            final_model = ModelClass(verbose=0, random_state=42, **best_params)
        else:
            final_model = ModelClass(**best_params)

        final_model.fit(X_train_data, y_train)

        # Сохранение результата
        BEST_RESULTS[f'{model_name}_{data_version}'] = {
            'model': final_model,
            'data_version': data_version,
            'best_params': best_params,
            'best_cv_rmse': best_cv_rmse,
            'X_train': X_train_data,
            'X_test': X_test_data
        }

        print(f"Лучший RMSE на CV: {best_cv_rmse:.4f}")
```

# 5. Вычисление метрик на новых данных

```python
def evaluate_model(model, X_train, X_test, y_train, y_test):
    # Предсказания
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Метрики на Train
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    
    # Метрики на Test
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
    mae_test = mean_absolute_error(y_test, y_test_pred)
    r2_test = r2_score(y_test, y_test_pred)
    
    return {
        'RMSE (Train)': rmse_train,
        'RMSE (Test)': rmse_test,
        'MAE (Test)': mae_test,
        'R^2 (Test)': r2_test
    }

# Формирование сводной таблицы
summary_data = []
for key, res in BEST_RESULTS.items():
    model_name, data_version = key.split('_', 1)
    
    metrics = evaluate_model(
        res['model'], 
        res['X_train'], 
        res['X_test'], 
        y_train, 
        y_test
    )
    
    summary_data.append({
        'Модель': model_name,
        'Данные': data_version,
        'RMSE (CV Optuna)': res['best_cv_rmse'],
        'RMSE (Train)': metrics['RMSE (Train)'],
        'RMSE (Test)': metrics['RMSE (Test)'],
        'MAE (Test)': metrics['MAE (Test)'],
        'R^2 (Test)': metrics['R^2 (Test)'],
        'Delta RMSE (Test-Train)': metrics['RMSE (Test)'] - metrics['RMSE (Train)']
    })

summary_df = pd.DataFrame(summary_data)

# Сортировка по лучшей тестовой метрике (минимальный RMSE)
summary_df_sorted = summary_df.sort_values(by='RMSE (Test)', ascending=True).round(4)

print("\nСводная Таблица Метрик")
print(summary_df_sorted.to_markdown(index=False))

# Определение лучшей модели
best_model_row = summary_df_sorted.iloc[0]

print("\nИтоговый Выбор Лучшей Модели:")
print(f"Лучшая модель: **{best_model_row['Модель']}** на данных **{best_model_row['Данные']}**")
print(f"RMSE (Test): **{best_model_row['RMSE (Test)']:.4f}**")
print(f"R^2 (Test): **{best_model_row['R^2 (Test)']:.4f}**")
```
